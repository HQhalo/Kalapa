<h1>Preprocessing Data</h1>

```python
NO_NULL = 47
def drop_null_row(dataset,):
    return dataset.loc[np.sum(dataset.isnull(),axis=1)<NO_NULL],dataset.loc[np.sum(dataset.isnull(),axis=1)>=NO_NULL]
    
```

```python
def combine_age(age1,age2,mean):
    if age1 == 0:
        if age2 == 0:
            return np.nan # all ages are null
        else:
            return age2
    else:
        if age2 == 0:
            return age1
        else: 
            if abs(age1 - age2) > 20:
                if min(age1,age2) < 50 and max(age1,age2) > 40:
                    return min(age1,age2)
            return int((age1+age2)/2)

def processing_age(data,MEDIAN = 32,BIN_AGE = 5):
    dataset = data.copy()
    age1 = dataset.age_source1.fillna(0)
    age2 = dataset.age_source2.fillna(0)

    age = []
    
    for a1,a2 in zip(age1,age2):
        age.append(combine_age(a1,a2,MEDIAN))
    dataset = dataset.drop(["age_source1","age_source2"],axis = 1)
    
    dataset["age"] = pd.cut(age,bins= list(range(0,101,BIN_AGE)),labels=list(range(0,int(100/BIN_AGE))))    
    
    return dataset            
```

```python
def processing_province(data):
#     data.province = data.province.fillna("khongco")
    data.province = data.province.map(lambda x: str(x).lower())
    data.replace(to_replace = "Tnh Ho Bnh",value = "Tnh Ha Bnh")
    
    
    return data
def processing_district(data):
#     data.district = data.district.fillna("khongco")
    data.district = data.district.map(lambda x: str(x).lower())
    
    return data
```

```python
def preprocessing_train(data):
    dataset = data.copy()
    dataset,dataset_null = drop_null_row(dataset)
    dataset = processing_age(dataset)
    dataset = processing_province(dataset)
    dataset = processing_district(dataset)
    return dataset,dataset_null

def preprocessing_test(data):
    dataset = data.copy()
    dataset = processing_age(dataset)
    dataset = processing_province(dataset)
    dataset = processing_district(dataset)
    return dataset
```

```python
train_proed,dataset_null = preprocessing_train(train)
test_proed = preprocessing_test(test)
```
